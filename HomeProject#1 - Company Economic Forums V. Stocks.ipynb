# Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
import numpy as np
import seaborn as sb
import csv
import urllib
import json
```

# Methods

```python
def decYears(date):
    values = date.split('-')
    year = int(values[0])
    month = int(values[1])
    day = int(values[2])
    monthVal = month - 1
    monthDays = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if year % 4 == 0:
        monthDays[1] = 29
    if year % 100 == 0:
        monthDays[1] = 28
    if year % 400 == 0:
        monthDays[1] = 29
    sum = 0.0000
    for i in range(monthVal):
        sum += monthDays[i]
    sum += day
    dec = sum/365
    return (year + dec)
```

# Import Company DataFrames

```python
apple = pd.read_csv('AAPL.csv', delimiter = ',', header = 0)
samsung = pd.read_csv('SMSG.csv', delimiter = ',', header = 0)
amazon = pd.read_csv('AMZN.csv', delimiter = ',', header = 0)
alphabet = pd.read_csv('ALPH.csv', delimiter = ',', header = 0)
microsoft = pd.read_csv('MSFT.csv', delimiter = ',', header = 0)
tesla = pd.read_csv('TSLA.csv', delimiter = ',', header = 0)
```

# Converting Company Data

```python
for i in range(len(apple['Date'])):
    apple['Date'][i] = decYears(apple['Date'][i])
    
for i in range(len(samsung['Date'])):
    samsung['Date'][i] = decYears(samsung['Date'][i])
    
for i in range(len(amazon['Date'])):
    amazon['Date'][i] = decYears(amazon['Date'][i])
    
for i in range(len(alphabet['Date'])):
    alphabet['Date'][i] = decYears(alphabet['Date'][i])
    
for i in range(len(microsoft['Date'])):
    microsoft['Date'][i] = decYears(microsoft['Date'][i])
    
for i in range(len(tesla['Date'])):
    tesla['Date'][i] = decYears(tesla['Date'][i])
    
companies = [apple, samsung, amazon, alphabet, microsoft, tesla]
```

# Graphing Company Data - Dates V. Stock (Close)

```python
for i in companies:
    plt.plot(i['Date'], i['Close'])
    plt.show()
```
